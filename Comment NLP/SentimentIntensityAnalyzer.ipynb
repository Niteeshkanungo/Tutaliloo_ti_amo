```python
import numpy as np
import pandas as pd
import math
filename = 'training_data.csv'
df = pd.read_csv(filename, delimiter = '\t')

x = df['comment'][:500]
y = df['score'][:500]
```

```python
df['neg'] = 0
df['neu'] = 0
df['pos'] = 0
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
for sentence in x:
    sid = SentimentIntensityAnalyzer()
    ss = sid.polarity_scores(sentence) 
    print(ss)
```
